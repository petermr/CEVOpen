##**Convert .rds files to Pandas**
To save a single R object in a file, we use .rds. While Pandas is a library used for data manipulation in Python.

The following code changes .rds files to Panads. This helps to run R files in Python using pyreadr.

```python
import pyreadr
result = pyreadr.read_r('/content/pfocr-curation-terpene-edition/pfocr_curating_terpene.rds') # also works for RData
print(result)
# done! 
# result is a dictionary where keys are the name of objects and the values python
# objects. In the case of Rds there is only one object with None as key
df = result[None]
```
